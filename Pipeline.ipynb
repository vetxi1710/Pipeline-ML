```python
from sklearn.pipeline import Pipeline
##feature Scaling
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
steps=[("standard_scler",StandardScaler()),
       ("classifer",LogisticRegression())]
```

```python
steps
```

```python
pipe=Pipeline(steps)
```

```python
#visualize Pipeline
from sklearn import set_config
set_config(display="diagram")
```

```python
pipe
```

```python
#create dataset
from sklearn.datasets import make_classification
x,y=make_classification(n_samples=100)
```

```python
x,y
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
len(x_train),len(x_test)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.decomposition import PCA
from sklearn.svm import SVC
```

```python
steps=[("scaling",StandardScaler()),
       ("PCA",PCA(n_components=3)),
       ("classifier",SVC())]
steps
```

```python
pipe2=Pipeline(steps)
pipe2
```

<h1>Complex examples of columns transformer</h1>

```python
#for numerix varaible to use impute
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
```

```python
import numpy as np
numeric_processor=Pipeline(
    steps=[("imputation_mean",SimpleImputer(missing_values=np.nan,strategy="mean")),
     ("scaler",StandardScaler())]
)
```

```python
numeric_processor
```

```python
#Catagorical processing pipeline
from sklearn.preprocessing import OneHotEncoder
categorical_processor=Pipeline(
    steps=[("imputation_constant",SimpleImputer(fill_value="missing",strategy="contant")),
     ("onehot",OneHotEncoder(handle_unknown="ignore"))]
)
```

```python
categorical_processor
```

```python
#combine processing technques
from sklearn.compose import ColumnTransformer
```

```python
preprocessor=ColumnTransformer(
    [("categorical",categorical_processor,["gender","city"]),
     ("numerical",numeric_processor,["age","height"])]
)
```

```python
preprocessor
```

```python
from sklearn.pipeline import make_pipeline
```

```python
make_pipeline(preprocessor,LogisticRegression())
# make_pipeline(preprocessor,SVC())
```
